```python
y=df["Moody's Long-Term Rating"]
```

```python
categories=df[["Cons. code","Indep. Ind."]]
```

```python
nums=df.drop(["Moody's Long-Term Rating","Cons. code","Indep. Ind."],axis=1)
```

```python
import numpy as np
nums=nums.replace("n.a.",np.float32(0.0))
nums=nums.replace(",","")
```

```python
nums=nums.astype("float32")
```

```python
categories=pd.get_dummies(categories)
```

```python
categories=categories.astype('float32')
```

```python
y_formed= y.map(mapping)
```

```python
y_formed
```

```python
merged_df = pd.concat([nums,categories], axis=1)
```

```python
merged_df
```

```python
from sklearn.preprocessing import StandardScaler

# DataFrameの全体を正規化
scaler = StandardScaler()
merged_df = pd.DataFrame(scaler.fit_transform(merged_df), columns=merged_df.columns)
```

```python
merged_df.dtypes
```

```python
learn=merged_df[:150]
test=merged_df[150:]
```

```python
laern_y=y_formed[:150].values
laern_x=learn.values
```

```python
np.mean(laern_y)
```

```python
test_y=y_formed[150:].values
test_x=test.values
```

```python
laern_x
```

```python
#成型官僚
#tensorflowで実験
```

```python
import tensorflow as tf
```

```python
# モデルの構築
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_dim=laern_x.shape[1]),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# モデルのコンパイル
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# モデルの学習
history = model.fit(laern_x, laern_y, epochs=5, batch_size=32, validation_split=0.1)

# テストデータでの評価
test_loss, test_mae = model.evaluate(test_x, test_y)
print('Test MAE:', test_mae)
```
